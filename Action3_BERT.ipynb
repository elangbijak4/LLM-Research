```python
import torch
from transformers import BertForQuestionAnswering, BertTokenizer
```

```python
# Muat model dan tokenizer BERT yang telah di-fine-tune
model_name = "bert-large-uncased-whole-word-masking-finetuned-squad"
model = BertForQuestionAnswering.from_pretrained(model_name)
tokenizer = BertTokenizer.from_pretrained(model_name)

def answer_question(question, context):
    # Tokenisasi input (pertanyaan dan konteks)
    inputs = tokenizer.encode_plus(question, context, add_special_tokens=True, return_tensors="pt")
    input_ids = inputs["input_ids"].tolist()[0]

    # Prediksi jawaban menggunakan model BERT
    outputs = model(**inputs)
    answer_start = torch.argmax(outputs.start_logits)
    answer_end = torch.argmax(outputs.end_logits) + 1

    # Konversi token jawaban kembali ke string
    answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(input_ids[answer_start:answer_end]))

    return answer
```

```python
# Contoh penggunaan
context = """
Perusahaan X menyediakan berbagai produk dan layanan termasuk software manajemen proyek, aplikasi mobile, dan layanan konsultasi IT. Produk kami dirancang untuk meningkatkan produktivitas dan efisiensi bisnis Anda. Tim ahli kami selalu siap membantu Anda dengan solusi yang disesuaikan dengan kebutuhan Anda.
"""

question = "Apa saja produk yang ditawarkan oleh Perusahaan X?"

answer = answer_question(question, context)
print("Pertanyaan:", question)
print("Jawaban:", answer)
```
